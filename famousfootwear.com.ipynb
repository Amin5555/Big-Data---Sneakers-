```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import requests

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
from datetime import datetime,date
today = datetime.now()
time_now = today.strftime("%Y-%m-%d %H")
```

```python
PATH = 'C:/Users/AMINE/OneDrive/Bureau/chromedriver.exe'
url1 = 'https://www.famousfootwear.com/nike-mens-shoes?icid=blp_nike0_topnv_menhttps://www.famousfootwear.com/nike-mens-shoes?icid=blp_nike0_topnv_men'
url2 = 'https://www.famousfootwear.com/womens-nike-shoes?icid=blp_nike0_topnv_women'
url = [url1,url2]
```

```python
filename = 'famousfootwear.com_{}'.format(time_now)+'.csv'
f = open(filename, 'w')
headers = 'Name;price;original_price;Brand;Date\n'
f.write(headers)


for url in url:
    driver = webdriver.Chrome(PATH)
    driver.get(url)
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html)
    containers = soup.findAll('span','runa_command')


    for container in containers :
        name = container['name']
        price = container['list_price']
        print(name,price)
        brand = 'famousfootwear.com'
        f.write(name + ";" + price + ";" + brand + ";" + time_now +"\n")
f.close()
```
